```python
#创建文件
def create__file(file_path,msg):
    f=open(file_path,'w+')
    f.write(msg)
    f.close
```

```python
#!/usr/bin/env python3
# coding: utf-8
import xlrd

file_path="/Users/taoxuefeng/Desktop/work/myself_sql/数据同步/ods改造/ods数据同步/"

# 打开excel文件，创建一个workbook对象,book对象也就是fruits.xlsx文件,表含有sheet名
rbook = xlrd.open_workbook('/Users/taoxuefeng/Desktop/work/myself_sql/数据同步/ods改造/业务数据库采集表清单.xlsx')

# sheets方法返回对象列表,[<xlrd.sheet.Sheet object at 0x103f147f0>]
rbook.sheets()

# xls默认有3个工作簿,Sheet1,Sheet2,Sheet3
rsheet = rbook.sheet_by_index(0)  # 取第一个工作簿

# 循环工作簿的所有行
for row in rsheet.get_rows():
    host = row[2]    #RDS实例
    db = row[3]      # 数据库名
    before = row[4]  #命名前缀
    table = row[5]   # 品名所在的列
    index = row[6]   # 品名所在的列
    user = row[7]   # 用户名
    password = row[8]   # 密码
    
    host_name = host.value     #rds实例
    db_name = db.value         #库名
    before_name = before.value #前缀
    table_name = table.value   #表名
    col_list = index.value     #字段名
    user_name = user.value     #用户名
    password_num = password.value     #密码
    
    
    # 打印
    if table_name == "表名":
        continue
    else:
        file_name = file_path + table_name + ".sh"
        msg = """
#!/bin/bash
host=""" + host_name + """
port=3306
username=""" + user_name + """
password=""" + password_num + """
opts="?tinyInt1isBit=false&zeroDateTimeBehavior=round&useSSL=false&autoReconnect=true"

db=""" + db_name + """
url=jdbc:mysql://$host:$port/$db${opts}
mysql_table=al_register_company_process
col_list=""" + col_list + """\n
hive_db=temp
hive_table='sap_'$mysql_table'_snap_dt'
hive_part=$(date +%Y%m%d)

sqoop import \
\n--connect $url \
\n--username $username \
\n--password $password \
\n--table $mysql_table \
\n--columns $col_list \
\n--direct \
\n--hive-import \
\n--hive-database $hive_db \
\n--hive-table $hive_table \
\n--hive-partition-key dt \
\n--hive-partition-value "$hive_part" \
\n--null-string '\\N' \
\n--null-non-string '\\N' \
\n--fields-terminated-by '^' \
\n--delete-target-dir \
\n--compression-codec org.apache.hadoop.io.compress.SnappyCodec \
\n-m 1"""
        create__file(file_name,msg)
        #print(file_name)
        #print(file_name+ "\n" + db_name + "\n" + msg)
```
